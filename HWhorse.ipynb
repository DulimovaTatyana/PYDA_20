Задание 3. Работа с пропусками      
Рассчитать количество пропусков для всех выбранных столбцов. 
Принять и обосновать решение о методе заполнения пропусков по каждому столбцу на основе рассчитанных статистик                
и возможной взаимосвязи значений в них. Сформировать датафрейм, в котором пропуски будут отсутствовать.

```python
(df.isna().mean() * 100).round(2)
#пропуски от 18 до 20% по категориям дыхание, температура конечностей, боль,рект.температура. 8% пропусков по категории пульс.
```

Так как пропуски в параметрах Хирургия и Исход всего 0,33% и это данные категориальные, то заполним пропуски в Хирургии и Исходе часто встречающимися значениями.

```python
df['Исход'].fillna(df['Исход'].mode()[0], inplace= True)
df['Хирургия'].fillna(df['Хирургия'].mode()[0], inplace= True)
```

```python
print(df.groupby('Исход')['пульс'].median()) 
```

Сгруппировав данные по исходу выдно, что в среднем выжившие лошади имели пульс 54, а все умершие 82-86.  
Cоответственно можно заменить пропуски в пульсе на основании данных исхода.

```python
df['пульс'].fillna(df.groupby('Исход')['пульс'].transform('median'), inplace=True)
```

Учащенное дыхание бывает при повышенном пульсе и соотвествено тоже косвенно зависит от исхода. Группируем по исходу и заменяем медианой пропуски в дыхании

```python
df['дыхание'].fillna(df.groupby('Исход')['дыхание'].transform('median'), inplace=True)
```

Скорее всего при повышенном сердцебиении и дыхании лошади испытывали боль. Заполним пропуски в боли данными на основе Исхода.

```python
df['боль'].fillna(df.groupby(['Исход'])['боль'].transform('median'), inplace=True)
```
