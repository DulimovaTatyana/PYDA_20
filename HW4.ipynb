#Домашнее задание к лекции Функции.
1. Пользователь по команде р - может узнать владельца документа по его номеру.
2. Пользователь по команде s -может узнать на какой полке храниться документ.
3. Пользователь по кманде l - может увидеть полную инфрмацию по всем документам.
4. Пользователь по команде as - может добавить новую полку.
5. Пользователь по команде ds - может удалить существующую полку (если она пустая).

```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
        '1': ['2207 876234', '11-2'],
        '2': ['10006'],
        '3': []
      }
def function(number):
    for doc_number in documents:
        if doc_number['number'] == number:
            print('Владелец документа', doc_number['name'])
            break
    else:
        if doc_number != number:
            print('Документ не найден в базе')
            
                
def function_s(number):
    flag = False
    for shelf_number, shelf_list in directories.items():
        if number in shelf_list:
            print('Документ храниться на полке:', shelf_number)
            return
        else:
            flag = True
    if flag:
        print('Документ ненайден в базе')
    return
        
def function_l():
    for element in documents:
        print('№: ', element['number'], 'тип: ', element['type'], 'владелец: ', element['name'], 'полка хранения: ', str(list(directories)).strip("[]").replace("'",""))
    return()

def function_as(number):
    Flag = False
    for direct in directories.keys():
        if direct == number:
            print("Такой номер полки есть. Все номера полок: ", str(list(directories)).strip("[]").replace("'",""))
            return ()
        else:
            Flag = True
            if Flag:
                directories.update({number:''})
                print('Такого номера полки нет, номер добавлен. Новые номера полок:', str(list(directories)).strip("[]").replace("'",""))
                return ()
            
def function_ds(number):
    flag = False
    for key, value in directories.items():
        if key != number:
            flag = True
        else:
            if len(value) == 0:
                directories.pop(key)
                print('Полка удалена. Текущий перечень полок:', str(list(directories)).strip("[]").replace("'",""))
                return()
            else:
                print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок:', str(list(directories)).strip("[]").replace("'",""))
                return()
    if flag:
        print('Такой полки не существует.Текущий перечень полок:', str(list(directories)).strip("[]").replace("'", ""))
        return()            
while True:
    command = input('Введите команду:')
    if command == 'p':
        function(input('Введите номер документа:'))
    elif command == 's':
        function_s(input('Введите номер документа:'))
    elif command == 'l':
        function_l()
    elif command == 'as':
        function_as(input('Введите номер полки:'))
    elif command == 'ds':
        function_ds(input('Введите номер полки:'))
    elif command == 'q':
        break
    else:
        print('Некорректный ввод')
```
